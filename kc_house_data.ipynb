```python
df.drop(['id','date'],axis='columns',inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(y=df['price'])
plt.show()
```

```python
df_1=df[df['bedrooms']<10]
df_1.head()
```

```python
df_2=df[df['price']<df['price'].quantile(0.99)]
df_2
```

```python
df_2['bathrooms']=df_2['bathrooms'].round().astype(int)
df_2
```

```python
df_2['bedrooms']=df_2['bedrooms'].round().astype(int)
```

```python
df_3=df_2[df_2['bathrooms']<=df_2['bedrooms']+1]
df_3
```

```python
df_4=df_3[df_3['sqft_living']<df_3['sqft_living'].quantile(0.99)]
```

```python
df_5=df_4[df_4['sqft_lot']<df_4['sqft_lot'].quantile(0.99)]
```

```python
df_5.drop(['sqft_living15','sqft_lot15'],axis='columns',inplace=True)
```

```python
df_5
```

```python
df_5.drop(['yr_renovated'],axis='columns',inplace=True)
df_5
```

```python
df_5['zipcode_encoded']=df_5.groupby('zipcode')['price'].transform('mean')
```

```python
df_5.drop(['zipcode'],axis='columns',inplace=True)
df_5
```

```python
plt.scatter(df_5['bedrooms'],df_5['price'],alpha=0.5)
plt.Figure(figsize=(10,6))
plt.xlabel("bedrooms")
plt.ylabel('price')
plt.grid()
plt.show()
```

```python
df_6=df_5[df_5['bedrooms']<=10]
```

```python
plt.scatter(df_6['bedrooms'],df_6['price'],alpha=0.5)
plt.Figure(figsize=(10,6))
plt.xlabel("bedrooms")
plt.ylabel('price')
plt.grid()
plt.show()
```

```python
plt.Figure(figsize=(10,6))
sns.stripplot(x=df_6['bedrooms'],y=df_6['price'],data=df_6,jitter=True)
plt.xlabel("bedrooms")
plt.ylabel('price')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sqft_living', y='price', data=df_6, alpha=0.5)
plt.title('sqft_living vs Price')
plt.xlabel('Living Area (sqft)')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='grade', y='price', data=df_6)
plt.title('Price Distribution by Grade')
plt.xlabel('Grade')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df_6.corr(numeric_only=True), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
df_7=df_6.copy()
```

```python
df_7.drop(['price'],axis='columns',inplace=True)
df_7
```

```python
x=df_7
y=df_6['price']
```

```python
x_1=x.astype('float32')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_1,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
model_1=RandomForestRegressor(n_estimators=400,max_depth=15,random_state=42,criterion='poisson')
model_1.fit(X_train,y_train)
```

```python
model_1.score(X_test,y_test)
```

```python
import pickle

with open('house_price_model.pkl', 'wb') as f:
    pickle.dump(model_1, f)
```
